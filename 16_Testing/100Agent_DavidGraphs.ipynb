```python
import pandas as pd
import pickle
import networkx as nx
```

```python
def extract_edgelist(G):
    nodes = sorted(list(G.nodes))
    edges = []
    
    for node in nodes:
        edges.append(sorted(list(G[node])))
        
    return dict(zip(nodes, edges))
```

```python
chain_graph = nx.Graph()
for i in range(1,13):
    chain_graph.add_edge(i-1,i)
    
complete_graph = nx.complete_graph(13)

graphs = [chain_graph, complete_graph]


edgelists = []
for G in graphs:
    edgelists.append(extract_edgelist(G))
    
with open('DavidPaperNets_13.pickle', 'wb') as f:
    pickle.dump(edgelists, f)    
```
