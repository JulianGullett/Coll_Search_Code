```python
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
test1d = np.random.normal(size=[1000000, 1])

hist, xbins = np.histogram(test1d, bins=1000)

plt.figure(figsize=(4,4))
sns.histplot(test1d)
plt.show()
plt.clf()

plt.figure(figsize=(4,4))
sns.histplot(hist, bins=20)
plt.show()
plt.clf()
```

```python
test2d = np.random.normal(size=[1000000, 2])

x,y = test2d[:,0], test2d[:,1]
hist, xbins, ybins = np.histogram2d(x, y, bins=1000)

xlabels = [f'{x:.2f}' for x in (xbins[:-1] + xbins[1:]) / 2]
ylabels = [f'{y:.2f}' for y in (ybins[:-1] + ybins[1:]) / 2]
plt.figure(figsize=(4,4))
sns.heatmap(hist.T, xticklabels=xlabels, yticklabels=ylabels,
            annot=True, fmt='.0f', cbar=False)
plt.show()
plt.clf()

plt.figure(figsize=(4,4))
sns.histplot(hist, bins=20)
plt.show()
plt.clf()
```
